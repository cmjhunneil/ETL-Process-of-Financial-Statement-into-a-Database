```python
import pandas as pd
```

```python
# Load the first sheet of the Excel file into a data frame
df = pd.read_excel('/home/cmjhunneil/Documents/Financial Statement(Dirty).xlsx', sheet_name='Income Statement')
```

```python
df
```

```python
pd.set_option('display.max_rows', None)
```

```python
# These are the columns where you want to transfer the values
columns = ['ttm', 'end_of_2022', 'end_of_2021', 'end_of_2020', 'end_of_2019']

# Start and end row for each block of rows to transpose
start_row = 0
end_row = 185

# Increment for each block of rows
increment = 6

for i in range(start_row, end_row+1, increment):
    # Transpose the values from rows i+1 to i+5 to the specified columns
    df.loc[i, columns] = df.loc[i+1:i+5, 'breakdown'].values

    # Drop the rows i+1 to i+5
    df = df.drop(range(i+1, i+6))
```

```python
df = df.reset_index(drop=True)
```

```python
df['ttm'] = df['ttm'].replace('-', 0)
```

```python
pd.options.display.float_format = '{:.0f}'.format
```

```python
df
```

```python
df.to_csv('cleaned_income_statement.csv', index=False)
```
